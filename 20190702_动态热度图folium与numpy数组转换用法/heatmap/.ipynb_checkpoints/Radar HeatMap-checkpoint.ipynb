# HeatMapWithTime Radar

In this example we show the most basic usage of the HeatMapWithTime plugin.

We generate a random set of points with lat/lon coordinates to draw on the map, and then move these points slowly in a random direction to simulate a time dimension. The points are aranged into a list of sets of data to draw.

```python
import folium
import folium.plugins as plugins
import numpy as np
import os
import time
```

```python
print("#################area split######################")
loc_downleft = [30.0, 106.0]#latitude,longitude
loc_upright = [40.0, 116.0]
row = 50
col = 50

def area_split(loc_downleft,loc_upright,row,col):
    location = np.zeros(shape=(row,col,2),dtype=float)# 3 mean [lat,lon,val]
    #print("location_init\n",location)
    row_length = (loc_upright[0] - loc_downleft[0])/row
    col_length = (loc_upright[1] - loc_downleft[1])/col
    print("row_length",row_length)
    print("col_length",col_length)
    for i in range(row) :
        for j in range(col) :
            # location[i, j, :] = [37.9, 118.35, 1]
            location[i,j] = [loc_upright[0]-(i+0.5)*row_length,loc_downleft[1]+(j+0.5)*col_length]
    return (location)
radar_location = area_split(loc_downleft,loc_upright,row,col)
#print("radar_location\n",radar_location)
print("radar_location[0,0,:] = \n",radar_location[0,0,:])
print("radar_location.shape ",radar_location.shape)

from PIL import Image
import h5py
import shutil

def trans_img_shapedata(img_dir,out_dir):
	fw = h5py.File(out_dir+'/'+fwname, 'w')
	area_file_list = os.listdir(img_dir)
	#print('area_file_list is ',area_file_list)
	#print('num of area_file =',len(area_file_list))
	for area_filename in area_file_list:
		temp = np.zeros([1,row, col], dtype=float)
		imgfile_path = ''
		imgfile_path = img_dir+'/'+area_filename
		#print ("%s has been find!"%imgfile_path)
		img_list = os.listdir(imgfile_path)
		img_list.sort()#img_list.sort(key=lambda x:int(x[4:])) #gt02.png pd16.png
		#print('img_list is ',img_list)
		#print('num of file =',len(img_list))
		for i,imgname in enumerate(img_list):
			img_path = ''
			img_path = imgfile_path+'/'+imgname
			img = Image.open(img_path)
			#print ("%s has been find!"%imgname)
			img_array = np.array(img)
			#print ("%s has been find!"%img_path)
			#print("img_array.shape",img_array.shape)
			#print("img_array.dtype", img_array.dtype)
			#img_array = (255.0-img_array)/255.0#归一化处理，（0-1）
			img_array = 1 - img_array/256.0#归一化处理，（0-1）
			#img_array = np.zeros([row, col], dtype=float) # np.zeros elesun
			#print ("img_array",img_array)
			#temp = np.concatenate((temp,img_array),axis=0)
			#temp = np.row_stack((temp,img_array))
			if i == 0 :
				temp = img_array[np.newaxis,]
			else :
				temp = np.concatenate((temp, img_array[np.newaxis,]), axis=0)
			#print("temp.shape",temp.shape)
		fw[area_filename] = temp
	fw.close()
	return (temp)

img_dir = './results/radar_predrnn_pp_radar500_train001'
out_dir = "./out"
fwname = "imgs.h5"
h = w = 50
if os.path.exists(out_dir):
    #os.remove(out_dir) #删除整个目录
    shutil.rmtree(out_dir) 
os.makedirs(out_dir)
radar_data = trans_img_shapedata(img_dir,out_dir)
print ("radar_data.shape is",radar_data.shape)
#print ("radar_data[0,:,:] = \n",radar_data[0,:,:])
#print ("radar_data[0,2,49] = \n",radar_data[0,2,49])
#print ("radar_data[1,20,25] = \n",radar_data[1,20,25])
print ("radar_data[2,11,22] = \n",radar_data[2,11,22])
print("#################array reshape######################")
#转换过程：h5 data(19, 50, 50) to (19,50*50,3),3 =(lat,lon,val)
radar_data = radar_data.reshape((19,50*50,1), order='C')
radar_location = radar_location.reshape((50*50,2), order='C')
#print("radar_location\n",radar_location)
temp = np.insert(radar_data, 0, radar_location[:,1], axis=2)
radar_move_data = np.insert(temp, 0, radar_location[:,0], axis=2)
print("radar_move_data.shape ",radar_move_data.shape)
data3 = radar_move_data[:13,] #elesun
print ("data3.shape ",data3.shape)
#print ("data3[0,0:50,:] \n",data3[0,0:50,:])
#print ("data3[0,149,2] \n",data3[0,149,2])
#print (np.argwhere(data3 == 0.2823529411764706))
#print (np.argwhere(data3 == 0.984313725490196))
#print(np.where(data3 == 0.2823529411764706))
#print ("data3[0,149,2] \n",data3(np.where(data3 == 0.2823529411764706)))
#print ("data3[2,572,2] \n",data3[2,572,2])# x,y,z[2,11,22] --> x,y*50+z,2 [2,11,2]
data3 = data3.tolist()
```

```python
#m = folium.Map([39.9, 116.35], zoom_start=11)#zoom_start small or big  tiles='stamentoner' #data1 data2
m = folium.Map([35, 110], zoom_start=5)#zoom_start small or big  tiles='stamentoner'  #data3
#hm = plugins.HeatMapWithTime(data2,radius=22) # data1 data2
hm = plugins.HeatMapWithTime(data3,radius=10,min_opacity=0.2, max_opacity=0.6, scale_radius=False) # data3
#hm = plugins.HeatMap(radar_move_data[9,].tolist(), min_opacity=0.5, radius=15)
hm.add_to(m)
m
```
