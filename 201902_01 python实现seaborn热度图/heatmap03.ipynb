```python
#coding=utf-8
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib as mpl
```

```python
uniform_data = np.random.rand(3, 3)#随机生成一个矩阵
heatmap = sns.heatmap(uniform_data)
plt.show()
```

```python
ax = sns.heatmap(uniform_data, vmin = 0.2, vmax = 0.5)#大于0.5的是同一个颜色 小于同理
plt.show()
```

```python
normal_data = np.random.randn(3, 3)
ax = sns.heatmap(normal_data, center = 0)#右边的颜色刻度条以0为中心
plt.show()
```

```python
flights = sns.load_dataset("flights")
print(flights.head())
```

```python
flights = flights.pivot("month", "year", "passengers")
print(flights)
```

```python
ax = sns.heatmap(flights)
plt.show()
```

```python
ax = sns.heatmap(flights, annot = True,fmt = "d")
plt.show()
```

```python
ax = sns.heatmap(flights, linewidths = 0.5)#间距
plt.show()
```

```python
ax = sns.heatmap(flights, cmap="YlGnBu")#换颜色
plt.show()
```
